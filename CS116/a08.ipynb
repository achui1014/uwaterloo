Write a function find_target that takes an mxn grid of numbers, where m is the number of rows and n is the number of columns, and returns True if target exists in the grid and False otherwise. In our grid, numbers are arranged in a descending manner, where each row is sorted in a descending order, and the first number in each row is smaller than the last number of the preceding row. Your solution must run in O(log mn) at worst. Note that m,n >= 0.

```python
def within_lst(L, i):
  '''
  Returns True if the integer i is within the range of the list of integers L,
  and False otherwise.
  
  within_lst: (listof Int) Int -> Bool
  '''
  min_lst = L[-1]
  max_lst = L[0]
  if (i >= min_lst) and (i <= max_lst):
    return True
  return False

def target_lst(L, target):
  '''
  Returns the sublist in L in which the integer target is within its range.
  Otherwise, an empty list is returned.
  
  target_lst: (listof (listof Int)) -> (listof Int)
  '''
  for i in range(len(L)):
    if within_lst(L[i], target):
      return L[i]
  return []

def find_target(L, target):
  '''
  Returns True if target is found within the grid of numbers L and False
  otherwise.
  
  find_target: (listof (listof Int)) Int -> Bool
  
  Examples:
    find_target([[60,30,34,30,23], [20,16,11,10], [7,5,3,1]], 30) => True
  '''
  if target_lst(L, target) != []:
    lst = target_lst(L, target)
    for i in range(len(lst)):
      if lst[i] == target:
        return True
  return False

L = [[60,34,30,23],[20,16,11,10],[7,5,3,1]]
find_target(L, 20)

L = [[5.2, 4.8], [4.4, 4.2], [2.9, 2.0]]
find_target(L, 4.2)
```
