```python
## ****************************************************************************
## Ashley Chui (21003047)
## CS 116, Winter 2024
## Assignment 09, Question 1
## ****************************************************************************
##


def print_fn(major, min_grade, overflow):
  print("Major | min grade | overflow") 
  print("{0:^6}|{1:^11}|{2:^6}".format(major, min_grade, overflow))
  print("----------------------------")  

def sort_alpha(spots_available):
  '''
  Returns the dictionary of spots_available with its key values (course names)
  sorted in alphabetical order.
  
  sort_alpha: (dictof Str Nat) -> (dictof Str Nat)
  '''
  courses = sorted(spots_available)
  spots = {}
  for i in courses:
    spots[i] = spots_available[i]
  return spots

def course_acceptance(major, spots, requests):
  '''
  Returns a list of the major, minimum grade, and the overflow of students
  that were not accepted in that order. This is based on the number of spots
  offered to the major and number of requests in each grade range.
  
  course_acceptance: Str Nat (dictof Str Nat) -> (list Str Nat Nat)
  '''
  grade_ranges = ['95-100', '90-95', '85-90', '80-85']
  grade_ind  = 0
  min_grade = grade_ranges[0][0:2]
  overflow = 0
  while (spots > 0) and (grade_ind < len(grade_ranges)):
    if requests[grade_ranges[grade_ind]] > spots:
      overflow = requests[grade_ranges[grade_ind]] - spots
      min_grade = grade_ranges[grade_ind][0:2]
      spots = 0
    elif requests[grade_ranges[grade_ind]] == 0:
      overflow = 0
    else:
      spots = spots - requests[grade_ranges[grade_ind]]
      min_grade = grade_ranges[grade_ind][0:2]
    grade_ind += 1
  return [major, min_grade, overflow]
  
def min_grade(spots_available, total_requests):
  '''
  Returns None and prints the minimum grade of acceptance of each major and
  the overflow of students. If a major is not listed in one of the dictionaries,
  but listed in the other, it is ignored.
  
  Effects: Prints to the screen.
  
  min_grade: (dictof Str Nat) (dictof Str (dictof Str Nat)) => None
  Requires:
   * The values associated with the keys in the dict spots_available > 0
   * In total_requests:
      * Keys begin with "requests_" followed by a major's name
      * All values associated with the keys are dictionaries 
        and consistently contains the keys ('95-100','90-95','85-90','80-85').
      * Each dict of requests must amount to 1 or more
  
  Examples:
    
  '''
  print("----------------------------")
  spots = sort_alpha(spots_available)
  for i in spots:
    rqst = "requests_" + i
    if rqst in total_requests:
      stats = course_acceptance(i, spots_available[i], total_requests[(rqst)])
      print_fn(stats[0], stats[1], stats[2])
```

```python
## ****************************************************************************
## Ashley Chui (21003047)
## CS 116, Winter 2024
## Assignment 09, Question 2
## ****************************************************************************
##


# CONSTANTS
MESSAGE = "request rejected: not enough available bikes"
BIKE_TYPE=["Mountain", "Road", "Hybrid", "Electric", "Folding"]
MAKE=["Trek", "Giant", "Liv", "Electra", "Momentum"]
MODEL=["Roscoe 7", "Slash", "Madone", "Emonda", "Trance Advanced", "Glory Advanced", "Townie", "Loft"]
WHEEL_SIZE=[12, 16, 20, 24, 26, 27.5, 29]
FRAME=["Steel", "Aluminum", "Titanium"]
COLORS = ["Black", "Burgundy", "Blue", "Green"]
STATUS = ["rented", "not rented", "lost"]
CONDITION = ["like new", "good", "bad"]


class Bike:
  def __init__(self, bike_type, make, model, wheel_size, frame, color, hourly_rate, 
                 deposit, condition="good", status="not rented"):
    self.bike_type = bike_type
    self.make = make
    self.model = model
    self.wheel_size = wheel_size
    self.frame = frame
    self.color = color
    self.hourly_rate = hourly_rate
    self.deposit = deposit
    self.condition = condition
    self.status = status
        
  def __repr__(self):
    return "\n+------------+------------+------------+------------+\
        \n|   Type     |    Make    |    Model   | Wheel Size |\
        \n+------------+------------+------------+------------+\
        \n|{0:^12}|{1:^12}|{2:^12}|{3:^12}|\
        \n+------------+------------+------------+------------+\
        \n|   Frame    |    Color   |    Rate    |   Deposit  |\
        \n+------------+------------+------------+------------+\
        \n|{4:^12}|{5:^12}|{6:^12}|{7:^12}|".format(self.bike_type, self.make, 
                                        self.model, self.wheel_size,
                                        self.frame, self.color, 
                                        self.hourly_rate, self.deposit)
  def update_status(self, new_status):
    self.status = new_status
    
  def update_condition(self, new_condition):
    self.condition = new_condition

class RentalManagement:
  def __init__(self, bikes):
    if bikes != []:
      self.bikes = bikes
    else:
      self.bikes = []
            
  def add_bike(self, bike):
    self.bikes.append(bike)
            
  def rent_bikes(self, quantity, bike_type, wheel_size, max_rate):
    '''
    Returns None and prints a message to the screen that says if the
    type, quantity, and specifications of bikes are available for rent.
    '''
    rentals = []
    for bike in self.bikes:
      if (bike.bike_type == bike_type) and (bike.wheel_size == wheel_size):
        if bike.hourly_rate <= max_rate:
          if bike.status == "not rented":
            rentals.append(bike)
    if len(rentals) > quantity:
      rentals = rentals [0:quantity]
      for i in rentals:
        i.update_status("rented")
      print("rented")
    elif len(rentals) == quantity:
      for i in rentals:
        i.update_status("rented")
      print("rented")
    else:
      print(MESSAGE)
```

```python
## ****************************************************************************
## Ashley Chui (21003047)
## CS 116, Winter 2024
## Assignment 09, Question 3
## ****************************************************************************
##


# CONSTANTS
MESSAGE = "No available flights from the entered departure city to the destination city"

class Flight:
  def __init__(self, airline, flight_num, departure_city, destination_city, 
              departure_time, arrival_time, duration, cost, seats_available, 
              status="on time"):
    self.airline = airline
    self.flight_num = flight_num
    self.departure_city = departure_city
    self.destination_city = destination_city
    self.departure_time = departure_time
    self.arrival_time = arrival_time  
    self.duration = duration
    self.cost = cost
    self.seats_available = seats_available
    self.status = status
        
  def __repr__(self):
    return "===============================\
        \nFlight {0}, operated by {1}\
        \n-------------------------------\
        \n{2} ---> {3}\n-------------------------------\
        \nDeparts: {4}\nArrives: {5}\nDurarion: {6}\
        \nSeats available: {7}\nEconomy price: {8}\
        \nBusiness class:{9}\
        \n-------------------------------\
        \nStatus: {10}\
        \n===============================\
        ".format(self.flight_num,self.airline,self.departure_city, 
                                    self.destination_city,self.departure_time,
                                    self.arrival_time,self.duration,
                                    self.seats_available,self.cost,
                                    self.cost*1.8,self.status)
        
  def print_info(self):
    print("===============================")
    print("Flight {0}, operated by {1}".format(self.flight_num,
                                                   self.airline))
    print("-------------------------------")
    print("{0} ---> {1}".format(self.departure_city, 
                                    self.destination_city))
    print("-------------------------------")
    print("Departs: {0}\nArrives: {1}".format(self.departure_time,
                                                  self.arrival_time))
    print("Durarion: {0}".format(self.duration))
    print("Seats available: {0}".format(self.seats_available))
    print("Economy price: {0}\nBusiness class: {1}".format(self.cost, 
                                                               self.cost*1.8))
    print("-------------------------------")
    print(f"Status: {self.status}")
    print("===============================")

  def update_seat(self, change):
    self.seats_available = self.seats_available + change
   
  def update_status(self, new_status):
    self.status = new_status

class TravelCompany:
  def __init__(self, flights):
    if len(flights) > 0:
      self.flights = flights
    else:
      self.flights = []

  def cheapest_flight(self, departure_city, destination_city):
    city_flights = []
    for i in self.flights:
      if i.departure_city == departure_city:
        if i.destination_city == destination_city:
          if i.status != "canceled":
            city_flights.append(i)
    if city_flights == []:
      return MESSAGE
    cheapest = city_flights[0]
    for i in city_flights:
      if i.cost < cheapest.cost:
        cheapest = i
    return cheapest 

  def shortest_flight(self, departure_city, destination_city):
    city_flights = []
    for i in self.flights:
      if i.departure_city == departure_city:
        if i.destination_city == destination_city:
          if i.status != "canceled":
            city_flights.append(i)
    if city_flights == []:
      return MESSAGE
    shortest = city_flights[0]
    for i in city_flights:
      if i.duration < shortest.duration:
        shortest = i
    return shortest

  def book_ticket(self, f_num):
    flight = None
    for i in self.flights:
      if i.flight_num == f_num:
        flight = i
    if flight == None:
      print("Flight {0} is not found!".format(f_num))
    elif flight.seats_available > 0:
      print("You have booked flight {0} successfully!".format(f_num))
      flight.update_seat(-1)
    else:
      print("No available seats on flight {0}!".format(f_num))
    
# Some objects for testing
flight1 = Flight("KLM", "KL0696", "Toronto", "Amsterdam", 
                 "21:15", "10:45", 7.5, 434.0, 12, status="canceled")
flight2 = Flight("KLM", "KL0797", "Toronto", "Cuba", 
                 "21:15", "10:45", 7.5, 242.0, 12, status="canceled")
flight3 = Flight("KLM", "KL0898", "Toronto", "London", 
                 "21:15", "10:45", 7.5, 470.0, 12, status="canceled")
flight4 = Flight("KLM", "KL0109", "Toronto", "Amsterdam", 
                 "21:15", "10:45", 7.5, 600.0, 12, status="canceled")
flight5 = Flight("KLM", "KL0119", "Toronto", "Amsterdam", 
                 "21:15", "10:45", 7, 800.0, 12, status="on time")
flight6 = Flight("KLM", "KL012", "Toronto", "Amsterdam", 
                 "21:15", "10:45", 7.5, 120.0, 12, status="on time")
flight7 = Flight("AC", "AC2224", "Toronto", "New York",
                 "06:30", "8:00", 1.5, 92.1, 20, status="on time")
flight8 = Flight("FL", "FL023", "Toronto", "New York",
                 "5:45", "7:15", 1.5, 92.1, 30, status="on time")

flight_group = [flight1, flight2, flight3, flight4, flight5, flight6]
Expedia = TravelCompany(flight_group)

flight_group1 = [flight7, flight8]
TripAdvisor = TravelCompany(flight_group)
```
